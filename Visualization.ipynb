```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Iris.csv'  # Update the path if necessary
iris_df = pd.read_csv(file_path)
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# Create histograms for numeric columns
numeric_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
# Set up the matplotlib figure for multiple subplots
plt.figure(figsize=(10, 8))
```

```python
# Loop through each numeric column and plot a histogram
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)  # Arrange subplots in a 2x2 grid
    sns.histplot(iris_df[column], kde=True, bins=15, color="skyblue")
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
```

```python
# Adjust the layout so plots don't overlap
plt.tight_layout()

# Show the plot
plt.show()
```
