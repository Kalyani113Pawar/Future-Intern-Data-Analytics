# Calculate Summary statistics

```python
import numpy as np
import pandas as pd
```

```python
iris=pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
# Function to calculate statistics
def calculate_statistics(df):
    statistics = {}
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        stats = {
            'mean': df[column].mean(),
            'median': df[column].median(),
            'std_dev': df[column].std(),
            'variance': df[column].var(),
            'min': df[column].min(),
            'max': df[column].max(),
            '25th_percentile': df[column].quantile(0.25),
            '50th_percentile': df[column].quantile(0.50),
            '75th_percentile': df[column].quantile(0.75)
        }
        statistics[column] = stats
    return statistics

# Calculate statistics for the Iris dataset
iris_statistics = calculate_statistics(iris)

# Print the statistics
for column, stats in iris_statistics.items():
    print(f"Statistics for {column}:")
    for stat_name, stat_value in stats.items():
        print(f"  {stat_name}: {stat_value}")
    print("\n")
```
